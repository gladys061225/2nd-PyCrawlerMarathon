## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path = "chromedriver")
browser.get("https://www.ettoday.net/news/news-list.htm") 
```

```python
import time

for i in range(150):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
from datetime import datetime

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

now = datetime.now()
time =  now.strftime("%Y/%m/%d 00:00")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text > time:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime
from datetime import timedelta

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

now = datetime.now()
twohago = now - timedelta(hours = 2)
twohago_time = twohago.strftime("%Y/%m/%d %H:%M")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text > twohago_time:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path = "chromedriver")
browser.get("https://www.ettoday.net/news/news-list.htm") 
```

```python
from datetime import datetime
from datetime import timedelta

now = datetime.now()
threedago = now - timedelta(days = 2)
threedago_y = threedago.strftime("%Y")
threedago_m = threedago.strftime("%m")
threedago_m_n0 = threedago_m.lstrip('0')
threedago_d = threedago.strftime("%d")
threedago_d_n0 = threedago_d.lstrip('0')
```

```python
selectSite = Select(browser.find_element_by_id("selY"))
selectSite.select_by_value(threedago_y)
selectYear = Select(browser.find_element_by_id("selM"))
selectYear.select_by_value(threedago_m_n0)
selectYear = Select(browser.find_element_by_id("selD"))
selectYear.select_by_value(threedago_d_n0)

browser.find_element_by_id('button').click()
```

```python
import time
for i in range(150):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
from datetime import datetime
from datetime import timedelta

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

threedago_3 = threedago.strftime("%Y/%m/%d 15:00")
threedago_5 = threedago.strftime("%Y/%m/%d 17:00")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text > threedago_3 and d.find(class_="date").text < threedago_5:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
