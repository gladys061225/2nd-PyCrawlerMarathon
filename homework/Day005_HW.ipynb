```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests

r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text

response
```

```python
import json
import pandas as pd

data = json.loads(response)
Titles = []
Head = []

for d in data:
    Titles.append(d['title'])
    
print("Article number = ", len(Titles))

Contents = pd.DataFrame(data)
cols = []
for col in Contents.columns:
    cols.append(col)
print(cols)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for content in data:
    print(content['title'], content['createdAt'], content['commentCount'], content['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
import numpy as np

comment_no = []
like_no =[]

for comment in data:
    comment_no.append(int(comment['commentCount']))
    like_no.append(int(comment['likeCount']))

print("平均留言人數 = ", np.mean(comment_no))
print("平均按讚人數 = ", np.mean(like_no))
```
