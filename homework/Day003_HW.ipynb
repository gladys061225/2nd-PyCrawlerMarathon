### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

with open('./Data/Day003/64_72hr_CH.xml',encoding="utf-8") as fd:
    doc = dict(xmltodict.parse(fd.read()))

n = 0
chapters = doc['cwbopendata']['dataset']['locations']['location']
for chapter in chapters:
    print (chapter['locationName'])
    n += 1
print("高雄市共有", n,'區')
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

info = doc['cwbopendata']['dataset']['locations']['location']
for i in info:
    print(i['locationName'],':',i['weatherElement'][0]['time'][0]['dataTime'],'-',i['weatherElement'][0]['time'][0]['elementValue']['value'],"度")
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

info_1 = doc['cwbopendata']['dataset']['locations']['location']
print(info_1[0]['locationName'])

for i in info_1[0]['weatherElement'][0]['time']:
    print(i['dataTime'],'-',i['elementValue']['value'],"度")
```
