```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests

r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text

response
```

```python
import json
import pandas as pd

data = json.loads(response)
Titles = []
Head = []

for d in data:
    Titles.append(d['title'])
    
print("Article number = ", len(Titles))

Contents = pd.DataFrame(data)
cols = []
for col in Contents.columns:
    cols.append(col)
print(cols)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for content in data:
    print(content['title'], content['createdAt'], content['commentCount'], content['likeCount'])
```
