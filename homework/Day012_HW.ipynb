### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ettoday.net/news/news-list.htm"

r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")
```

```python
for articles in soup.find(class_="part_list_2").find_all("h3"):
    print(articles.find(class_="date").text, articles.find_all("a")[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd

c = []
a = []
for t in soup.find(class_= "part_list_2").find_all("h3"):
    c.append(t.find_all("em")[-1].text)
    a.append(t.find_all("a")[-1].text)
    
cat = pd.DataFrame({"article": a}, index = c)
print(cat.loc["社會"])
```

### ③ 哪一個類別的文章最多

```python
category = list(set(c))
for i in category:
    print(i,":", c.count(i))
```
